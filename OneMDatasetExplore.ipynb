```python
def organize_data(row):
    r = {}
    row_lst = row[0].split('::')
    
    r['user_id'] = row_lst[0]
    if row_lst[1] == 'M':
        r['gender'] = 0
    else:
        r['gender'] = 1

    
    age = int(row_lst[2])
    age_ = [0 for _ in range(7)]
    if age == 1:
        age_[0] = 1
    elif age == 18:
        age_[1] = 1
    elif age == 25:
        age_[2] = 1
    elif age == 35:
        age_[3] = 1
    elif age == 45:
        age_[4] = 1
    elif age == 50:
        age_[5] = 1
    elif age == 56:
        age_[6] = 1
    else:
        pass
    r['age']= age_
        
	# 25:  "25-34"
	# 35:  "35-44"
	# 45:  "45-49"
	# 50:  "50-55"
	# 56:  "56+"

    

    occupation = [0 for _ in range(20)]
    if row_lst[3] == '0':
        pass
    else:
        occupation[int(row_lst[3])-1] = 1
            
    r['occupation'] = occupation

    location = [0 for _ in range(11)]
    zip_code = row_lst[4]
    #print(location)
    canada = False
    for k in zip_code:
        if k in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ':
            canada = True
    
    if canada:
        location[10] = 1
    else:
        location[int(zip_code[0])] = 1
    r['zip_code'] = location

    return r
```

```python
Data = []
for row in users.values:
    Data.append(organize_data(row))
```

```python
pd.DataFrame(Data).to_csv(r"C:\Users\yyBee\Datasets\ml-1m\users.csv")
```

```python
import chardet


# Read the file as binary to detect encoding
with open(r"C:\Users\yyBee\Datasets\ml-1m\movies.dat", 'rb') as f:
    result = chardet.detect(f.read())
    encoding = result['encoding']

# Read the CSV with the detected encoding
items = pd.read_csv(r"C:\Users\yyBee\Datasets\ml-1m\movies.dat", encoding=encoding, delimiter='::', header = None)
print(items)
```

```python
items.columns = ['movie_id','title','genre']
```

```python
items
```

```python
genre_2_indices = {'Action':0, 'Adventure':1, 'Animation':2, "Children's":3, 'Comedy':4 ,
	'Crime':5, 'Documentary':6, 'Drama':7,  'Fantasy':8, 'Film-Noir':9, 'Horror': 10,
	'Musical':11, 'Mystery':12, 'Romance':13, 'Sci-Fi':14, 'Thriller':15, 'War':16, 'Western':17}


def process_items(row):

    genres = row['genre'].split('|')
    encoded = np.zeros((18,))
    for g in genres:
        idx = genre_2_indices[g]
        encoded[idx] = 1
    row['genre'] = list(encoded)
    return row
    
```

```python
items.apply(process_items, axis = 1)
```
